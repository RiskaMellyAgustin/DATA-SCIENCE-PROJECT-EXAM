```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("D:/1.UNI 5th Semester/6. Advance DataBase/ADVANCE_DB PROJECT/ds_project/mellyapp/ml/mellyapp_dataset_OLAP.csv")
df.head()
```

```python
print(df.info())
print("\nDeskripsi Statistik:")
print(df.describe())
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['grade'], kde=True, bins=20, color='skyblue')
plt.title("Distribusi Nilai Akhir Mahasiswa")
plt.xlabel("Grade")
plt.ylabel("Jumlah Mahasiswa")
plt.tight_layout()
plt.show()
```

```python
# --- 4. Korelasi antar Fitur ---
plt.figure(figsize=(10, 8))
corr = df.drop(['stu_id', 'course_id'], axis=1).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Heatmap Korelasi antar Fitur")
plt.tight_layout()
plt.show()
```

```python
# --- 5. Boxplot untuk Outlier ---
num_cols = ['age', 'total_activities', 'total_duration_minutes',
            'quiz_count', 'individual_assignment_count',
            'group_assignment_count', 'forum_count', 'grade']
for col in num_cols:
    plt.figure(figsize=(6, 3))
    sns.boxplot(x=df[col], color='lightgreen')
    plt.title(f"Boxplot: {col}")
    plt.tight_layout()
    plt.show()
```

```python
# --- 6. Distribusi Grade per Gender ---
plt.figure(figsize=(6, 4))
sns.boxplot(x='gender', y='grade', data=df)
plt.title("Distribusi Grade per Gender")
plt.tight_layout()
plt.show()
```

```python
# --- 7. Scatter Plot Aktivitas vs Grade ---
plt.figure(figsize=(6, 4))
sns.scatterplot(x='total_activities', y='grade', hue='gender', data=df)
plt.title("Total Aktivitas vs Grade")
plt.tight_layout()
plt.show()
```
